```python
df = pd.read_csv('../input/where-is-time-spent-by-the-people-around-the-world/Time-Use-in-OECD-Countries-OECD.csv')
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,15))
sns.scatterplot(data=df, x='Time (minutes)', y='Country', palette='deep', hue='Category', size='Category', sizes=(20,100))
plt.legend(loc = 2, bbox_to_anchor = (1,1))
plt.grid(axis='y')
```

* Japan spends the most time in minutes on paid work, sleeps the least, spends the least time caring for households.  
* Latvia on second place in the time on paid work. Together with Lithuania spends a minimum of time in minutes for education.  
* Italia spends the minimum time in minutes on paid work.  
* South Africa spends the most time in minutes on sleep.  
* Korea spends the most time in minutes on Education.
* Mexico spends the most time in minutes on houswork.
* India and South Africa spend almost no time on shopping
* Most people spend their time sleeping.  
  
And you can see other categories in which people from different countries spend their time (in minutes).
